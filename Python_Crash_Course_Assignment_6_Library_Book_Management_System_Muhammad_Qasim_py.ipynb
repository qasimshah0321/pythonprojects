```python
books = {
    "Python Crash Course": {"author": "Eric Matthes", "available": True},
    "Fluent Python": {"author": "Luciano Ramalho", "available": True},
    "Effective Python": {"author": "Brett Slatkin", "available": True},
    "Learning Python": {"author": "Mark Lutz", "available": True}

}

borrowed_books = []

def view_books():
    print("\n Available Books:")
    for title, info in books.items():
        status = "Available" if info["available"] else "Not Available"
        print(f'"{title}" by {info["author"]} — {status}')
    print()

def borrow_book():
    book_title = input("Enter the book name to borrow: ").strip().title()
    if book_title in books:
        if books[book_title]["available"]:
            books[book_title]["available"] = False
            borrowed_books.append(book_title)
            print(f' You have borrowed "{book_title}".')
        else:
            print(f' "{book_title}" is currently not available.')
    else:
        print(" Book not found in the library.")

def return_book():
    book_title = input("Enter the book name to return: ").strip().title()
    if book_title in borrowed_books:
        books[book_title]["available"] = True
        borrowed_books.remove(book_title)
        print(f' You have returned "{book_title}".')
    else:
        print(" You have not borrowed this book.")

def add_book():
    title = input("Enter new book title: ").strip().title()
    author = input("Enter author name: ").strip().title()
    if title in books:
        print(" Book already exists in the catalog.")
    else:
        books[title] = {"author": author, "available": True}
        print(f' Book "{title}" by {author} added to the library.')

def view_borrowed_books():
    print("\n Borrowed Books:")
    if borrowed_books:
        for book in borrowed_books:
            print(f'- {book}')
    else:
        print("No books have been borrowed.")
    print()

while True:
    print("\n Welcome to the Library System")
    print("1. View Books")
    print("2. Borrow Book")
    print("3. Return Book")
    print("4. Add a New Book")
    print("5. View Borrowed Books")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        view_books()
    elif choice == "2":
        borrow_book()
    elif choice == "3":
        return_book()
    elif choice == "4":
        add_book()
    elif choice == "5":
        view_borrowed_books()
    elif choice == "6":
        print(" Exiting the system. Goodbye!")
        break
    else:
        print(" Invalid choice. Please select between 1–6.")
```
