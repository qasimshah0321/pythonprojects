```python
expenses = {}

def add_expense():
    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category: ").capitalize()
    try:
        amount = int(input("Enter amount: "))
    except ValueError:
        print("Invalid amount. Please enter a number.")
        return

    if date not in expenses:
        expenses[date] = []
    expenses[date].append({"category": category, "amount": amount})
    print("Expense added")

def view_all_expenses():
    if not expenses:
        print("No expenses recorded yet.")
        return
    print("\nAll Expenses:")
    for date, records in expenses.items():
        print(f"\nDate: {date}")
        for entry in records:
            print(f"  - {entry['category']}: Rs. {entry['amount']}")

def view_by_category():
    category = input("Enter category to search: ").capitalize()
    total = 0
    for date in expenses:
        for entry in expenses[date]:
            if entry["category"] == category:
                total += entry["amount"]
    print(f"Total spent on {category}: Rs. {total}")

def view_by_date():
    date = input("Enter date (YYYY-MM-DD): ")
    if date in expenses:
        print(f"\nExpenses on {date}:")
        for entry in expenses[date]:
            print(f"  - {entry['category']}: Rs. {entry['amount']}")
    else:
        print("No expenses found for this date.")

# Main loop
while True:
    print("\nDaily Expense Tracker")
    print("1. Add Expense")
    print("2. View All Expenses")
    print("3. View by Category")
    print("4. View by Date")
    print("5. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_all_expenses()
    elif choice == "3":
        view_by_category()
    elif choice == "4":
        view_by_date()
    elif choice == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid option. Please try again.")
```
